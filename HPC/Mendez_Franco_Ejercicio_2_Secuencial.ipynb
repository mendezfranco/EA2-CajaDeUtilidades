**3. Desarrollo**

A continuación se codifican las funcionalidades para cumplir con el objetivo del ejercicio.

```
# --------------------------------------------
#@title 3.1 Parámetros de ejecución { vertical-output: true }

tamanio_matriz = 50#@param {type: "number"}

# --------------------------------------------

from datetime import datetime

tiempo_total = datetime.now()

import numpy as np
# --------------------------------------------
# Definición de función que transforma el tiempo en  milisegundos 
tiempo_en_ms = lambda dt:(dt.days * 24 * 60 * 60 + dt.seconds) * 1000 + dt.microseconds / 1000.0

# --------------------------------------------

# CPU - Defino la matriz de NxN, asignándole valores aleatorios comprendidos 
# entre -50 y 50
matriz_cpu = np.random.uniform(low=-50.0, high=50.0, size=(tamanio_matriz,tamanio_matriz))#.random((tamanio_matriz,tamanio_matriz))

# CPU - Defino la matriz de NxN resultante, que obtendrá el producto de elevar
# la matriz original al cuadrado
matriz_resultante_cpu = np.zeros_like(matriz_cpu)

# CPU - Inicializo las variables de recorrido
i=0
j=0
k=0
suma=0

tiempo_bucle = datetime.now()

for i in range(tamanio_matriz):
   for j in range(tamanio_matriz):
       for k in range(tamanio_matriz):
           suma += matriz_cpu[i][k] * matriz_cpu[k][j]
       matriz_resultante_cpu[i][j] = suma
       suma=0

tiempo_bucle = datetime.now() - tiempo_bucle

tiempo_total = datetime.now() - tiempo_total

print("Matriz original: ")
print(matriz_cpu)
print("")

print("Matriz al cuadrado: ")
print(matriz_resultante_cpu)
print("")

print("Tiempos Totales: ")
print("Tiempo Bucle: ", tiempo_en_ms(tiempo_bucle))
print("Tiempo Total: ", tiempo_en_ms(tiempo_total))



```

**5. Conclusiones**

*Repaso*

La potencia de matrices, y por consiguiente la multiplicación de matrices, conlleva un proceso con elevada carga de trabajo, que deriva en una complejidad de procesamiento que aumenta proporcionalmente a la cantidad de filas-columnas que posee la matriz.
Como podemos observar, el algoritmo deberá efectuar una cantidad de operaciones que puede llegar a ser elevada. Al realizarse estas operaciones de manera secuencial, el tiempo que demorará obtener la matriz resultado (matriz elevada al cuadrado) será notablemente mayor comparado con una solución en paralelo, ya que se deberán recorrer cada fila y columna dato a dato, para luego almacenar el resultado en una nueva matriz. La cantidad de operaciones que se van a llevar a cabo aumentan exponencialmente conforme aumente el tamaño de la matriz, siguiendo la relación *Cantidad =*image.png. Así, por ejemplo, para una matriz de 10x10, se harán un total de 1000 cálculos matemáticos **secuenciales**. Esta secuencialidad puede derivar en tiempos totals y de respuesta de larga duración, afectando en la performance del algoritmo.
